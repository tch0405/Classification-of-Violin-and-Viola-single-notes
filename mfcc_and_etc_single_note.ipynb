```python
import os
import numpy, scipy, matplotlib.pyplot as plt, IPython.display as ipd
import librosa, librosa.display
import numpy as np
import math
os.chdir('/Users/tanchonghong/Documents/final_github1/viola')
x_train1=[]
y_train1=[]
mfcc_list_mean = []
mfcc_list_std = []
count1=0
for f_name in os.listdir('/Users/tanchonghong/Documents/final_github1/viola'):
    if f_name.endswith('.wav') or f_name.endswith('.mp3'):
        x, sr = librosa.load(f_name)
        
        start=0
        end = 4*22050

        x = x/x.max()
        length = len(x)
        
            
        mfcc = librosa.feature.mfcc(x, n_mfcc=13)
        variable = []
        temp1 =list(np.mean(mfcc,1))
        temp2=list(np.std(mfcc,1))
        variable = temp1+temp2
        
        
        variable.append(np.mean(librosa.feature.zero_crossing_rate(x)))
        variable.append(np.mean(librosa.feature.spectral_bandwidth(y=x, sr=sr)))
        variable.append(np.mean(librosa.feature.spectral_centroid(y=x, sr=sr)))
        variable.append(np.mean(librosa.feature.spectral_rolloff(y=x, sr=sr)))
        variable.append(np.mean(librosa.feature.spectral_flatness(y=x)))
        x_train1.append(variable)
        y_train1.append("viola")
            
os.chdir('/Users/tanchonghong/Documents/final_github1/violin')
for f_name in os.listdir('/Users/tanchonghong/Documents/final_github1/violin'):
    if f_name.endswith('.wav') or f_name.endswith('.mp3'):
        x, sr = librosa.load(f_name)
        start=0
        end = 4*22050

        x = x/x.max()
        mfcc = librosa.feature.mfcc(x, n_mfcc=13)
        variable = []
        temp1 =list(np.mean(mfcc,1))
        temp2=list(np.std(mfcc,1))
        variable = temp1+temp2
        variable.append(np.mean(librosa.feature.zero_crossing_rate(x)))
        variable.append(np.mean(librosa.feature.spectral_bandwidth(y=x, sr=sr)))
        variable.append(np.mean(librosa.feature.spectral_centroid(y=x, sr=sr)))
        variable.append(np.mean(librosa.feature.spectral_rolloff(y=x, sr=sr)))
        variable.append(np.mean(librosa.feature.spectral_flatness(y=x)))
        x_train1.append(variable)
        y_train1.append("violin")
        
        
        
```

```python
from numpy.random import seed
np.random.seed(10)
seed(10) 
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x_train1, y_train1, test_size = 0.30)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
#scaler = StandardScaler()
#scaler.fit(x_train)
#x_train = scaler.transform(x_train)
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
classifier = RandomForestClassifier(max_depth=12, random_state=0)
#classifier = SVC(kernel='rbf', random_state=0, gamma=.1, C=1,probability = True)

classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
y_pred = classifier.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print('Accuracy' + str(accuracy_score(y_test, y_pred)))
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test,y_pred))
```
